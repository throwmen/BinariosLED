# If Statement

```
  <code>
...

  <code>
```


  


````python
age = 15

if  age >= 18:
  print('u r stupid')
else:
  print("I'm starting to get bored")
````

# For loop



```
for <variable> in <list>:
  <code>
```



````python
countries
````

````python
for i, country in enumerate(countries):
  print(i)
  print(country)
````

````python
my_data = {'name': 'Frank', 'age': 26}
my_data
````

````python
for key, value in my_data.items():
  print(key)
  print(value)
````
